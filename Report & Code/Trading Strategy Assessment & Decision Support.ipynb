```python
# Load and concat data
files = [
    "Replay_Trading_OANDA_US30USD_2025-07-17.xlsx",
    "Replay_Trading_OANDA_US30USD_2025-07-19.xlsx",
    "Replay_Trading_OANDA_US30USD_2025-07-21.xlsx",
    "Replay_Trading_OANDA_US30USD_2025-07-22.xlsx",
    "Replay_Trading_OANDA_US30USD_2025-07-23.xlsx",
    "Replay_Trading_OANDA_US30USD_2025-07-25.xlsx"
]

sheet_name = "List of trades"

# Load and label each sheet
dfs = []
for i, file in enumerate(files, start=1):
    df_temp = pd.read_excel(file, sheet_name=sheet_name)
    df_temp["Source File"] = f"Sheet{i}" 
    dfs.append(df_temp)

df = pd.concat(dfs, ignore_index=True)

df["Trade #"] = pd.to_numeric(df["Trade #"], errors="coerce")

# Show data
df.head(5)
```

```python
# Drop Cumulative P&L USD and Cumulative P&L %	as they reset based on each Sheet Source 
df.drop(columns=["Cumulative P&L USD","Cumulative P&L %"],inplace=True)
```

```python
# Calculate total return% based on the exit P&L%
exit_rows = df[df["Type"].str.contains("Exit")]

cumulative_return = 1.0
cumulative_pct = []


for pnl_pct in exit_rows['P&L %']:
    cumulative_return *= (1 + pnl_pct)  
    cumulative_pct.append((cumulative_return - 1) * 100)


round(cumulative_pct[-1],2)
print(f"Total Return: {cumulative_pct[-1]:.2f}%")
```

```python
# Calculate the average trade P&L%
Average_Trade_Performace = df["P&L %"].mean() * 100
print(f"Average Trade Performace: {Average_Trade_Performace:.2f}%")
```

```python
# Extract Market time columns from Date/time to end up with date/time(jo) and data/time (makret time)
utc_plus_3 = pytz.FixedOffset(180)

all_trades['Date/Time_Entry'] = pd.to_datetime(all_trades['Date/Time_Entry']).dt.tz_localize(utc_plus_3)
all_trades['Date/Time_Exit'] = pd.to_datetime(all_trades['Date/Time_Exit']).dt.tz_localize(utc_plus_3)

all_trades['Date/Time_Entry (Market Time)'] = all_trades['Date/Time_Entry'].dt.tz_convert('US/Eastern')
all_trades['Date/Time_Exit (Market Time)'] = all_trades['Date/Time_Exit'].dt.tz_convert('US/Eastern')

all_trades.rename(columns={'Date/Time_Entry': 'Date/Time_Entry (JO)','Date/Time_Exit': 'Date/Time_Exit (JO)'}, inplace=True)

for col in ['Date/Time_Entry (JO)', 'Date/Time_Exit (JO)', 'Date/Time_Entry (Market Time)', 'Date/Time_Exit (Market Time)']:
    all_trades[col] = all_trades[col].dt.tz_localize(None)

all_trades
```

```python
# Create trade hold duration in hours column 
all_trades['Holding Duration (hrs)'] = (all_trades['Date/Time_Exit (JO)'] - all_trades['Date/Time_Entry (JO)']).dt.total_seconds() / 3600

# Create outcome column win/lose
all_trades["Trade Outcome"] = all_trades["P&L USD"].apply(lambda x: "Win" if x > 0 else "Loss")

all_trades
```

```python
# Create a new column "Risk Zone" based on the hold duration , first get the 75th percentile of hold duration for losing trades to act as a danger threshold 
# and the average hold duration for the winning trades to act as safe threshhold and what is between these 2 times is Caution (the process will be sperated by Dircetion)

long_trades = all_trades[all_trades["Direction"] == "Long"]
short_trades = all_trades[all_trades["Direction"] == "Short"]

danger_hold_long = long_trades[long_trades['P&L USD'] < 0]['Holding Duration (hrs)'].quantile(0.75)
danger_hold_short = short_trades[short_trades['P&L USD'] < 0]['Holding Duration (hrs)'].quantile(0.75)

safe_hold_long = long_trades[long_trades['P&L USD'] > 0]['Holding Duration (hrs)'].mean()
safe_hold_short = short_trades[short_trades['P&L USD'] > 0]['Holding Duration (hrs)'].mean()

def classify_risk_zone(row):
    if row['Direction'] == 'Long':
        if row['Holding Duration (hrs)'] <= safe_hold_long:
            return 'Safe'
        elif row['Holding Duration (hrs)'] <= danger_hold_long:
            return 'Caution'
        else:
            return 'Danger'
    elif row['Direction'] == 'Short':
        if row['Holding Duration (hrs)'] <= safe_hold_long:
            return 'Safe'
        elif row['Holding Duration (hrs)'] <= danger_hold_short:
            return 'Caution'
        else:
            return 'Danger'
    return 'Unknown'


all_trades['Risk Zone'] = all_trades.apply(classify_risk_zone, axis=1)
all_trades.head(2)
```

```python
# Define ET market sessions
def get_market_session(hour, minute):
    time = hour + minute / 60
    if 20 <= time or time < 4: 
        return 'Overnight (20:00 - 04:00 ET)'
    elif 4 <= time < 9.5:
        return 'Pre-Market (04:00 - 09:30 ET)'
    elif 9.5 <= time < 10.5:
        return 'First Hour (09:30 - 10:30 ET)'
    elif 10.5 <= time < 15:
        return 'Midday-Afternoon (10:30 - 15:00 ET)'
    elif 15 <= time < 16:
        return 'Last Hour (15:00 - 16:00 ET)'
    elif 16 <= time < 20:
        return 'After-Hours (16:00 - 20:00 ET)'
    else:
        return 'Overnight (20:00 - 04:00 ET)'

session_order = [
    'Overnight (20:00 - 04:00 ET)',
    'Pre-Market (04:00 - 09:30 ET)',
    'First Hour (09:30 - 10:30 ET)',
    'Midday-Afternoon (10:30 - 15:00 ET)',
    'Last Hour (15:00 - 16:00 ET)',
    'After-Hours (16:00 - 20:00 ET)'
]

all_trades['Date/Time_Entry (Market Time)'] = pd.to_datetime(all_trades['Date/Time_Entry (Market Time)'])
all_trades['Date/Time_Exit (Market Time)'] = pd.to_datetime(all_trades['Date/Time_Exit (Market Time)'])

all_trades['Entry_Session_ET'] = all_trades['Date/Time_Entry (Market Time)'].apply(lambda x: get_market_session(x.hour, x.minute))
all_trades['Exit_Session_ET'] = all_trades['Date/Time_Exit (Market Time)'].apply(lambda x: get_market_session(x.hour, x.minute))

all_trades['Entry_Session_ET'] = pd.Categorical(all_trades['Entry_Session_ET'], categories=session_order, ordered=True)
all_trades['Exit_Session_ET'] = pd.Categorical(all_trades['Exit_Session_ET'], categories=session_order, ordered=True)
```
